# IMPLEMENTATION OF DECISION TREE USING REFINED ATTRIBUTES AS FEATURES AND PRICE AS THE OTCOME

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df3  = pd.read_csv("housingnew.csv")
```

```python
df3.head()
```

```python
y = df3["price"]
```

```python
x = df3.drop(["price","date"],axis = 1)
```

```python
x.columns
```

```python
x.shape
```

```python
x.describe()
```

```python
y.shape
```

```python

from sklearn import tree
from sklearn.tree import DecisionTreeRegressor  
regressor = DecisionTreeRegressor(random_state = 0) 
regressor.fit(x, y) 
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=7)
```

```python
regressor.fit(x_train,y_train)
```

```python
regressor.score(x_test,y_test)
```

```python
tree.plot_tree(regressor.fit(x_train,y_train))
```

```python
from sklearn.metrics import mean_squared_error
```

```python
regressor.predict([[7,3,2,1000,2000,2,3,7,1000,1000,2014,47.5112,-122.257,1340]])
```

```python
mse = mean_squared_error(y_test, regressor.predict(x_test))
np.sqrt(mse)
```
